```python
def rand_plot(N,NN2): ## NN2 here is the class variable!!!
    for j in range(N):
        for i in range(len(NN2.W)): ## set the W and b are all  generate random networks 
            r,c = NN2.W[i].shape;
            NN2.W[i] = np.random.randn(r,c);
            NN2.b[i] = np.random.randn(c); 
            
            #print(np.mean(NN2.W[i]),np.std(NN2.W[i]))
            #print(np.mean(NN2.b[i]),np.std(NN2.b[i]))
        
        
        q2 = NN2.Forward((NN2.L + 1), Xgrid, NN2.W, NN2.b); ## the last node is the y for plot
        Y = q2[-1][:,0]; ## Y is a matrix rather than an array, so need to get it 0th column to plot
        fig = plt.figure();
        ax = fig.add_subplot(1,1,1,projection = '3d');
        #ax.scatter(Xgrid[:,0],Xgrid[:,1],Y)
        ax.plot_trisurf(Xgrid[:,0],Xgrid[:,1],Y,linewidth = 0.2, antialiased = True,color = 'Yellow');       
        
rand_plot(8,NN2);
```
